```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('/scratch/zt1/project/msml612/user/jeffy22/MSML612_EnergyForecasting/PatchTST/PatchTST_dual_scale/results/ETTh1_336_96_dual_best_combo_DualScalePatchTST_ETTh1_ftM_sl336_ll48_pl96_dm128_nh8_el3_dl1_df256_fc1_ebtimeF_dtTrue_test_0/losses.csv')

# Plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

epochs = range(1, len(df) + 1)

ax1.plot(epochs, df['train_loss'], 'b-', linewidth=2)
ax1.set_ylabel('Loss (MSE)')
ax1.set_title('Training Loss')
ax1.grid(True, alpha=0.3)

ax2.plot(epochs, df['valid_loss'], 'r-', linewidth=2)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss (MSE)')
ax2.set_title('Validation Loss')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
